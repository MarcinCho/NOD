```python
# Grupowanie wierszy według wartości kolumny kategoryzowanej, potem uśrednienie wartości wszystkich kolumn w grupie MultiIndex
outcome = df.groupby(['ethnicity','age']).mean(numeric_only=True)
outcome
```

```python
# Grupowanie wierszy według wartości kolumny kategoryzowanej, potem uśrednienie wartości dla pewnych kolumn, liczba wartości i mediana dla pozostałych kolumn w grupach
agregations = {
    'wmn_hh' : 'mean',
    'wmn_pre_iud_missed' : 'median'
}
outcome = df.groupby('age').agg(agregations)
outcome
```

```python
# Stworzenie tabeli przestawnej (pivot table) na podstawie ramki danych

tabela_przestawna = pd.pivot_table(
    data = df,
    index = 'geography',
    columns='age',
    values = 'ethnicity',
    aggfunc= 'sum'

)
tabela_przestawna
```

```python
# Wyświetlenie indeksów i kolumn tabeli przestawnej
tabela_przestawna.index
```

```python
# Utworzenie indeksu złożonego tabeli przestawnej i wyświetlenie go
tabela_przestawna_2 = pd.pivot_table(
    data = df,
    index = ['geography','sex'],
    columns='age',
    values = 'wmn_hh',
    aggfunc= 'sum'

)
tabela_przestawna_2
```

```python
# Przedstawienie sposobów łączenia ramek danych za pomocą metod merge i concat

df_pracownicy = pd.DataFrame({'ID_Pracownika': [1, 2, 3], 'Imię': ['Anna', 'Piotr', 'Kasia']})
df_dzialy = pd.DataFrame({'ID_Pracownika': [2, 3, 4], 'Dział': ['IT', 'HR', 'Marketing']})

# Złączenie wewnętrzne (inner join) na podstawie 'ID_Pracownika'
wynik_merge = pd.merge(df_pracownicy, df_dzialy, on='ID_Pracownika', how='inner')
```

```python

# Pokazanie dodawania nowych kolumn za pomocą operacji matematycznych
df = pd.DataFrame({
    'Cena': [100, 150, 200],
    'Ilość': [5, 3, 2]
})

# Dodanie nowej kolumny 'Przychod'
df['Przychod'] = df['Cena'] * df['Ilość']

df
```

```python
# Przedstawienie na przykładzie dodawania nowych kolumn z pomocą funkcji lambda

df = pd.DataFrame({
    'Przychód': [1500, 2200, 900, 3100],
    'Koszt': [1000, 1500, 1200, 2500]
})
df['Wskaźnik_Zysku'] = df.apply(
    lambda x: (x['Przychód'] - x['Koszt']) / x['Koszt'],
    axis=1  # KLUCZOWE: Uruchamia funkcję wzdłuż wierszy
)
df
```
